## 1. Compare Students' engagement data based on class factor

```python
import numpy as np

def linegraph_draw(data, label, color):
    y, binEdges = np.histogram(data, bins=20)
    bincenters = 0.5 * (binEdges[1:] + binEdges[:-1])
    plt.plot(bincenters, y, '-', c=color, label=label)
```

### 1-1. Num of Logins

```python
plt.rcParams["figure.figsize"] = (12,6)

plt.hist(df_0['# Logins'], bins=20)
plt.axis([min(df['# Logins']), max(df['# Logins']), 0, 225])
plt.show()
```

```python
np.std(df_0['# Logins'])
```

```python
plt.hist(df_1['# Logins'], bins=20)
plt.axis([min(df['# Logins']), max(df['# Logins']), 0, 225])
plt.show()
```

```python
np.std(df_1['# Logins'])
```

```python
plt.hist(df_2['# Logins'], bins=20)
plt.axis([min(df['# Logins']), max(df['# Logins']), 0, 225])
plt.show()
```

```python
np.std(df_2['# Logins'])
```

```python
linegraph_draw(df_0['# Logins'], "Non Pass", "g")
linegraph_draw(df_1['# Logins'], "Fair", "r")
linegraph_draw(df_2['# Logins'], "Good", "b")

plt.legend()
plt.show
```
